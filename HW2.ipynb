```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

lr_model = LinearRegression(normalize=True)
```

```python
x = np.random.rand(200, 1) * 20
y = (x - 4)*(x - 6)*(x - 14)*(x - 16)
#y = (x + 8)*(x +1)*(x - 10)*(x - 25)
y = y + np.random.normal(scale=3, size=y.shape)

x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
split = int(x.shape[0] * 0.8)

x_train = x[:split]
y_train = y[:split]

x_test = x[split:]
y_test = y[split:]
```

```python
MSE_train = []
MSE_test = []
```

```python
lr_model.fit(x_train, y_train)

x_line = np.arange(0, 20, 0.1).reshape(-1, 1)
y_line = lr_model.predict(x_line)

plt.plot(x_train, y_train, 'bo')
plt.plot(x_test, y_test, 'co')
plt.plot(x_line, y_line, 'r--')
```

```python
h_train = lr_model.predict(x_train)
mse_train = np.mean((y_train - h_train) ** 2) / 2

h_test = lr_model.predict(x_test)
mse_test = np.mean((y_test - h_test) ** 2) / 2

MSE_train.append(mse_train)
MSE_test.append(mse_test)
```

```python
x_train = np.hstack((x_train, x_train**2))

lr_model.fit(x_train, y_train)

x_line = np.arange(0, 20, 0.1).reshape(-1, 1)
x_line = np.hstack((x_line, x_line**2))
y_line = lr_model.predict(x_line)

plt.plot(x_train[:, 0], y_train, 'bo')
plt.plot(x_test, y_test, 'co')
plt.plot(x_line[:, 0], y_line, 'r--')
```

```python
x_test = np.hstack((x_test, x_test**2))

h_train = lr_model.predict(x_train)
mse_train = np.mean((y_train - h_train) ** 2) / 2

h_test = lr_model.predict(x_test)
mse_test = np.mean((y_test - h_test) ** 2) / 2

MSE_train.append(mse_train)
MSE_test.append(mse_test)
```

```python
x_test = x_test[:, 0].reshape(-1, 1)
x_train = x_train[:, 0].reshape(-1, 1)
x_train = np.hstack((x_train, x_train**2, x_train**3, x_train**4))

lr_model.fit(x_train, y_train)

x_line = np.arange(0, 20, 0.1).reshape(-1, 1)
x_line = np.hstack((x_line, x_line**2, x_line**3, x_line**4))
y_line = lr_model.predict(x_line)

plt.plot(x_train[:, 0], y_train, 'bo')
plt.plot(x_test, y_test, 'co')
plt.plot(x_line[:, 0], y_line, 'r--')
```

```python
x_test = x_test[:, 0].reshape(-1, 1)
x_test = np.hstack((x_test, x_test**2, x_test**3, x_test**4))

h_train = lr_model.predict(x_train)
mse_train = np.mean((y_train - h_train) ** 2) / 2

h_test = lr_model.predict(x_test)
mse_test = np.mean((y_test - h_test) ** 2) / 2

MSE_train.append(mse_train)
MSE_test.append(mse_test)
```

```python
x_test = x_test[:, 0].reshape(-1, 1)
x_train = x_train[:, 0].reshape(-1, 1)
for i in range(2,9):
    z = x_train[:, 0].reshape(-1, 1)
    x_train = np.hstack((x_train, z**i))
    
lr_model.fit(x_train, y_train)

x_line = np.arange(0, 20, 0.1).reshape(-1, 1)
for i in range(2,9):
    z_line = x_line[:, 0].reshape(-1, 1)
    x_line = np.hstack((x_line, z_line**i))

y_line = lr_model.predict(x_line)

plt.plot(x_train[:, 0], y_train, 'bo')
plt.plot(x_test, y_test, 'co')
plt.plot(x_line[:, 0], y_line, 'r--')
```

```python
x_test = x_test[:, 0].reshape(-1, 1)
for i in range(2,9):
    z = x_test[:, 0].reshape(-1, 1)
    x_test = np.hstack((x_test, z**i))

h_train = lr_model.predict(x_train)
mse_train = np.mean((y_train - h_train) ** 2) / 2

h_test = lr_model.predict(x_test)
mse_test = np.mean((y_test - h_test) ** 2) / 2

MSE_train.append(mse_train)
MSE_test.append(mse_test)
```

```python
x_test = x_test[:, 0].reshape(-1, 1)
x_train = x_train[:, 0].reshape(-1, 1)
for i in range(2,17):
    z = x_train[:, 0].reshape(-1, 1)
    x_train = np.hstack((x_train, z**i))
    
lr_model.fit(x_train, y_train)

x_line = np.arange(0, 20, 0.1).reshape(-1, 1)
for i in range(2,17):
    z_line = x_line[:, 0].reshape(-1, 1)
    x_line = np.hstack((x_line, z_line**i))

y_line = lr_model.predict(x_line)

plt.plot(x_train[:, 0], y_train, 'bo')
plt.plot(x_test, y_test, 'co')
plt.plot(x_line[:, 0], y_line, 'r--')
```

```python
x_test = x_test[:, 0].reshape(-1, 1)
for i in range(2,17):
    z = x_test[:, 0].reshape(-1, 1)
    x_test = np.hstack((x_test, z**i))
    
h_train = lr_model.predict(x_train)
mse_train = np.mean((y_train - h_train) ** 2) / 2

h_test = lr_model.predict(x_test)
mse_test = np.mean((y_test - h_test) ** 2) / 2

MSE_train.append(mse_train)
MSE_test.append(mse_test)
```
